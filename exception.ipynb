```python
# ZeroDivisionError
try:
  a=5/0
  print(a)
except ZeroDivisionError:
  print("division by zero in not applicabe")
```

```python
# ValueError
try:
  num=int(input("Enter the number : "))
  print (10/num)
except ValueError:
  print("error: invalid value")
```

```python
try:
  num=int(input("Enter the number :"))
  print(10/num)
except ZeroDivisionError:
  print("error: division by zero is not applicable")
except ValueError:
  print("error: invalid value")
```

```python
try:
  num1=int(input("enter the first number :"))
  num2=int(input("enter the second number :"))
  print(num1/num2)
except (ZeroDivisionError,ValueError):
  print("error: check the inputs and give a valid value")
```

```python
#Raise
age=int(input("enter the age :"))
if age<18:
  raise ValueError("age can not be below 18")
else:
  print("Eligible")
```

```python
# Type error
try:
  print(5+"s")
except TypeError:
  print("error : invalid value, enter number value")
```

```python
# index
try:
  course=["python","power bi","excel","SQL"]
  print(course[6])
except IndexError:
  print("error : index not found")
```

```python
# keyError
try:
  details={"name":"sezhiyan","age":"21","gender":"male"}
  print(details["native"])
except KeyError:
  print("Not found give valid details")
```
